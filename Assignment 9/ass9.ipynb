```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.dropna(axis=0, inplace=True)
```

```python
sns.distplot(titanic.age)
```

```python
sns.boxplot(x='sex', y = 'age', data = titanic, hue = 'survived')
```

```python
sns.scatterplot(x='sex', y='age', data=titanic, hue='survived')
plt.title = "Distribution of age with respect to gender along with survived"
plt.xlable = "Gender"
plt.xlable = "Age"
plt.show()
```

```python
titanic.survived.value_counts().plot(kind='pie')
```

```python
plt.hist(titanic.age, bins=20)
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=titanic, hue='survived')
```

```python
sns.heatmap(pd.crosstab(titanic.pclass, titanic.age))
```
