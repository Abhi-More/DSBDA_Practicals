```python
#import all dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#load inbuild 'titanic' datasets
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
print("Shape: ", titanic.shape)
print("Data types: ")
titanic.dtypes
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic.dropna(axis='index', inplace=True)
```

```python
#Distributional Plot
sns.distplot(titanic['fare'])
```

```python
#The Joint Plot
sns.jointplot(x='age', y='fare', data=titanic, hue='sex')
```

```python
sns.jointplot(x='age', y='fare', data=titanic, kind='hex')
```

```python
#pairplot
sns.pairplot(titanic, hue='sex')
```

```python
#rugplot
sns.rugplot(titanic.fare)
```

```python
sns.barplot(x='sex', y='age', data=titanic)
```

```python
#Calculate the Aggregate values of category using estimate

sns.barplot(x='sex', y='age', data=titanic, estimator = np.std)
```

```python
sns.boxplot(x='sex', y='age', data=titanic)
```

```python
sns.barplot(x='pclass', y='survived', data=titanic, hue="sex")
```

```python
sns.stripplot(x='sex', y='fare', data = titanic, hue='sex')
```

```python
sns.stripplot(x='sex', y='age', data = titanic, hue='sex', jitter = True)
```

```python
sns.violinplot(x='sex', y='age', data=titanic, hue='sex')
```

```python
sns.swarmplot(x='sex', y='fare', data = titanic, hue = 'survived')
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.show()
```
