# Different Way to load dataframe

```python
# Convert the dictionary into DataFrame

data = {
        "Sr. No.":[1, 2, 3, 4, 5, 6],
        "Name" :["Abhishek", "Disha", "Darshan", "Revati", "Chetan", "Sakshi"],
        "Relation":[4, 5, np.nan, 7,8,9]
    }
  
df = pd.DataFrame(data)
  
df.isnull()
```

```python
#loading list into dataframe

list = ["Abhishek", "Saish", "Sandip", "Darshan", "Ganesh", "Ashutosh"]
list2 = [23, 342, 423, 54, 54, 56]
df = pd.DataFrame(list, list2)
df
```
