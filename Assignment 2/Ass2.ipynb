dropna() method to deal with missing values

```python
#droping the rows with missing values
df1 = df.dropna(axis=0)
df1.isnull().sum()

#df.dropna(axis, how, thresh, subset, inplace)
```

```python
#droping column with any missing value
df2 = df.dropna(axis=1, how='any')
df2.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df3 = df.dropna(axis='columns', thresh=471)
df3.isnull().sum()
```

fillna() method

```python
#filling missing values with single value
df["gender"] = df["gender"].fillna("Male")
df["gender"].isnull().sum()
```

```python
#filling missing values from previous record
df["raisedhands"] = df.raisedhands.fillna(method='pad')
df["raisedhands"].isnull().sum()
```

```python
#filling missing values from next record
df["VisITedResources"] = df.VisITedResources.fillna(method='bfill')
df["VisITedResources"].isnull().sum()
```

```python
#filling missing values in dataframe itself
df["AnnouncementsView"].fillna(method='pad', inplace=True)
df["AnnouncementsView"].isnull().sum()
```

```python
df.fillna(method='pad', inplace=True)
df.isnull().sum()
```
